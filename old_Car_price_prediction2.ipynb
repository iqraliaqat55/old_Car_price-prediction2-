
#Preprocessing a Dataset for Machine Learning: Encoding and Scaling

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
dataset
data = {
    'car_name': ['Jeep Compass 2.0 Longitude Option BSIV', 'Honda City VX', 'Hyundai Creta'],
    'car_prices_in_rupee': [1500000, 900000, 1200000],
    'kms_driven': [50000, 30000, 40000],
    'fuel_type': ['Diesel', 'Petrol', 'Diesel'],
    'transmission': ['Manual', 'Automatic', 'Manual'],
    'ownership': ['First', 'Second', 'First'],
    'manufacture': [2019, 2018, 2020],
    'engine': [2000, 1500, 1600],
    'Seats': [5, 5, 5],
}

# Create DataFrame
dataset = pd.DataFrame(data)

# Separate target and input data
output_data = dataset['car_prices_in_rupee']
input_data = dataset.drop(['car_prices_in_rupee', 'car_name'], axis=1)

# Encode categorical data
input_data = pd.get_dummies(input_data, drop_first=True)

# Scale numeric data
ss = StandardScaler()
input_data_scaled = ss.fit_transform(input_data)
input_data = pd.DataFrame(input_data_scaled, columns=input_data.columns)

# Output
print(input_data)
```
